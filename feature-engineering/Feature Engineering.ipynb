```python
# Required libraries
import json
import pandas as pd
from pandas import read_csv
import re

specs_df = pd.read_csv("all_brand_models_specs.csv")
```

```python
# reading the JSON data using json.load()
file = 'all_brand_models_specs.json'
with open(file) as train_file:
    dict_train = json.load(train_file)

# converting json dataset from dictionary to dataframe
train = pd.DataFrame.from_dict(dict_train, orient='index')
train.reset_index(level=0, inplace=True)
```

```python
print(dict_train['Samsung']['Galaxy S9+']['Network'])
print(len(dict_train['Samsung']['Galaxy S9+']))
for var in dict_train['Samsung']['Galaxy S9+']:
    print(var)
    
# wanted nested keys
#['Network']['Technology'] -> check for GSM/HSPA/LTE (binary - 1 if yes, 0 if no)
#['Body']['Dimension'] -> check for dimensions (calculate, split by x)
#['Body']['Weight'] -> in grams
#['Display']['Type'] -> type of screen
#['Display']['Size'] -> area
#['Display']['Resolution'] -> split by ppi
#['Memory']['Card slot'] -> Binary (1 if yes, 0 if no)
#['Memory']['Internal']
#['Camera']['Video']
#['Camera']['Primary'] -> check for 'Dual'
#['Comms']['Radio'] -> binary
#['Tests']['Battery life']
#['Features']['Sensors'] -> check for 'Face ID'
```

```python
# Getting all phone brands
brands = []
for key in dict_train:
    brands.append(key)

# Getting all phone models
models = []
all_brands = []
for brand in brands:
    for key in dict_train[brand]:
        all_brands.append(brand)
        models.append(key)
        
# Getting all gsm network
gsm_arr = []
hspa_arr = []
lte_arr = []
for brand in brands:
    for model in models:
        try:
            if ('GSM' in dict_train[brand][model]['Network']['Technology']):
                gsm_arr.append(1)
            else:
                gsm_arr.append(0)
        except KeyError:
            continue
# Getting all hspa network
for brand in brands:
    for model in models:
        try:
            if ('HSPA' in dict_train[brand][model]['Network']['Technology']):
                hspa_arr.append(1)
            else:
                hspa_arr.append(0)
        except KeyError:
            continue
# Getting all lte network            
for brand in brands:
    for model in models:
        try:
            if ('LTE' in dict_train[brand][model]['Network']['Technology']):
                lte_arr.append(1)
            else:
                lte_arr.append(0)
        except KeyError:
            continue
print(len(gsm_arr))
print(len(hspa_arr))
print(len(lte_arr))
```

```python
#print(brands)
#print(models)
#print(dict_train['Apple']['iPhone X']['Body'])
dimensions = []
# Getting all dimensions(incomplete)         
for brand in brands:
    for model in models:
        try:
            str_dimension = dict_train[brand][model]['Body']['Dimensions']
            #print(str_dimension)
            if ("(" in str_dimension):
                str_dimension = str_dimension.split("(")[1]
            if 'oz' in str_dimension:
                str_dimension = str_dimension.split(" ")[0]
                dimensions.append(float(str_dimension))
            if 'in)' in str_dimension:
                str_dimension = ((str_dimension.replace("in)", "")).replace(" ", "")).split("x")
                vol = 1
                for i in str_dimension:
                    vol *= float(i)
                vol = round(vol/1.8,2)
                dimensions.append(vol)
        except ValueError:
            #print(brand)
            #print(model)
            #print(num)
            print()
        except KeyError:
            continue
average = sum(dimensions)/len(dimensions)
for i in range(40):
    dimensions.append(average)
```

```python
# Getting all weights
weight_arr = []
for brand in brands:
    for model in models:
        try:
            weight = dict_train[brand][model]['Body']['Weight']
            cleaned_weight = ((weight.replace("g", "")).split("/")[0]).split(";")[0]
            cleaned_weight = (re.sub(r"(.*?)\s?\(.*?\)", r"\1", cleaned_weight).lower()).replace(" ", "")
            cleaned_weight = cleaned_weight.split(".")[0]
            cleaned_weight = re.sub("\D", "", cleaned_weight)
            if (cleaned_weight == ""):
                cleaned_weight = 0
            weight_arr.append(int(cleaned_weight))
        except KeyError:
            continue
average = round(sum(weight_arr)/len(weight_arr))
# Assign 0 weights with average weight
for weight in range(len(weight_arr)):
    if weight_arr[weight] == 0:
        weight_arr[weight] = average
print(len(weight_arr))
```

```python
# Getting display type (incomplete)
display_arr = []
for brand in brands:
    for model in models:
        try:
            display = dict_train[brand][model]['Display']['Type']
            print(display)
            #if "capacitive" not in display:
                #print(brand)
                #print(model)
                #print(display)
             #   cleaned_display = display.split("capacitive")[0]
                #print(cleaned_display)
        except KeyError:
            continue
```

```python
# Getting screen size
# 1 inch is 96 pixels/2.54cm
# 1 cm is 37.8 pixels
# 1cm2 is 0.155inch2
screensize_arr = []
for brand in brands:
    for model in models:
        try:
            size = dict_train[brand][model]['Display']['Size']
            if "ppi" in  size:
                size = (size.split("~")[1]).split(" ")[0]
            print(size)
        except KeyError:
            continue
```

```python
# Getting memory card slot
slot_arr = []
for brand in brands:
    for model in models:
        try:
            slot = dict_train[brand][model]['Memory']['Card slot']
            if slot == 'No' or slot == " ":
                slot = 0
            elif "+" in slot:
                slot = 0
            elif slot == "To be confirmed":
                slot = 0
            else:
                if len(slot.split(",")) > 2:
                    slot = slot.split(",")[1]
                slot = (re.sub(r"(.*?)\s?\(.*?\)", r"\1", slot)).split("-")[0]
                slot = re.sub("[^0-9]", "", slot)
                if slot == "":
                    slot = 0
            slot_arr.append(int(slot))
        except KeyError:
            continue
print(len(slot_arr))
```

```python
# Getting memory internal
count = 0
internal_arr = []
for brand in brands:
    for model in models:
        try:
            slot = dict_train[brand][model]['Memory']['Internal']
            count += 1
            slot = ((((((slot.split("-")[0]).split("or")[0]).split("(")[0]).split(",")[0]).replace(" RAM", "")).split("+")[0]).replace(" ROM", "")
            size = ""
            if len(slot.split(" ")) > 1:
                size += slot.split(" ")[1]
                slot = (slot.split(" ")[0]).split("/")[0]
                slot += size
            internal_arr.append(slot)
        except KeyError:
            continue

for i in range(14):
    internal_arr.append('4GB')
```

```python
# Getting camera video
video_1080 = []
video_2160 = []
video_720 = []
video_480 = []
dual_video = []
for brand in brands:
    for model in models:
        try:
            video = dict_train[brand][model]['Camera']['Video']
            if "2160p@" in video:
                video_2160.append(1)
            elif "2160p@" not in video:
                video_2160.append(0)
            else:
                video_2160.append(0)
            if "1080p@" in video:
                video_1080.append(1)
            elif "1080p@" not in video:
                video_1080.append(0)
            else:
                video_1080.append(0)
            if "720p@" in video:
                video_720.append(1)
            elif "720p@" not in video:
                video_720.append(0)
            if "480p@" in video:
                video_480.append(1)
            elif "480p@" not in video:
                video_480.append(0)
            if "dual-video" in video:
                dual_video.append(1)
            elif "dual-video" not in video:
                dual_video.append(0)
        except KeyError:
            continue

for i in range(12):
    video_1080.append(0)
    video_720.append(0)
    video_480.append(0)
    video_2160.append(0)
    dual_video.append(0)

print(len(video_1080))
print(len(video_720))
print(len(video_480))
print(len(video_2160))
print(len(dual_video))
```

```python
#['Comms']['Radio'] -> binary
#['Tests']['Battery life']
#['Features']['Sensors'] -> check for 'Face ID'

# Getting radio
radio_arr = []
for brand in brands:
    for model in models:
        try:
            radio = dict_train[brand][model]['Comms']['Radio']
            if (radio == "No"):
                radio = 0
            else:
                radio = 1
            radio_arr.append(radio)
        except KeyError:
            continue
for i in range(10):
    radio_arr.append(0)
print(len(radio_arr))
```

```python
# Getting battery
battery_arr = []
for brand in brands:
    for model in models:
        try:
            batt = dict_train[brand][model]['Battery']['Stand-by']
            batt = (re.sub(r"(.*?)\s?\(.*?\)", r"\1", batt)).split("/")[0]
            if "Up" not in batt:
                batt = ""
            batt = re.sub("[^0-9]", "", batt).replace(" ", "")
            if batt == '':
                batt = '0'
            battery_arr.append(int(batt))
        except KeyError:
            continue
print(len(battery_arr))
```

```python
d = {'brand': all_brands, 'model': models, 'network_GSM': gsm_arr ,'network_HSPA': hspa_arr, 'network_LTE': lte_arr,
     'weight': weight_arr, 'memory_card_slot': slot_arr, 'internal_memory': internal_arr, 'dual_camera': dual_video,
    '480_camera': video_480, "720_camera": video_720, "1080_camera": video_1080, "2160_camera": video_2160, 'radio': radio_arr,
    'dimensions(in oz)': dimensions}
tentative_cols = ['brand', 'model', 'network_GSM','network_HSPA', 'network_LTE', 'dimensions(in oz)', 'weight',
        'memory_card_slot', 'internal_memory', 'dual_camera', '480_camera', '720_camera', '1080_camera', '2160_camera',
        'radio']
cols = ['brand', 'model', 'network_GSM','network_HSPA', 'network_LTE', 'dimensions(in oz)', 'weight', 'display_type', 'screen_size', 'resolution_ppi',
        'memory_card_slot', 'internal_memory', 'video', 'dual_camera', '480_camera', '720_camera', '1080_camera', '2160_camera',
        'radio', 'battery_life']
feature_engineering_df = pd.DataFrame(data=d, columns=tentative_cols)
print(len(feature_engineering_df))
print(feature_engineering_df.dtypes)
feature_engineering_df.head()
```
